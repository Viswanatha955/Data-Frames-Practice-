# Merge data  

```python
movies = pd.DataFrame({'movieid':[1,2,3,4,5],
                     'title': ['Toy Story2','Mission Impossible','Avatar','Blade Runner','Ghost']
                     })
movies
```

```python
members = pd.DataFrame({'memid':[1,2,3,4],
                     'name': ['Tom','Jon','Aron','Bobby'],
                      'movieid': [1,3,4,np.NaN]
                     })
members
```

```python
pd.merge(movies,members,on='movieid')
```

```python
pd.merge(movies,members, on='movieid', how ='left')
```

```python
merged_df = pd.merge(movies,members, on ='movieid', how='right')
merged_df
```

```python
merged_df[['name','title']]
```

```python
# Select only 3 columns from the merged dataframe 
pd.merge(movies,members, on='movieid', how='right')[['name','title']]
```

```python
pd.merge(movies,members,on='movieid',how='outer')
```

## Handling Duplicates

```python
ddup=DataFrame({'col1':['A','A','B','B','B'], 'col2':[2,2,2,3,3]})
ddup
```

```python
#Use duplicated to find duplicates
ddup.duplicated()
```

```python
ddup.drop_duplicates()
```

```python
ddup.drop_duplicates(keep='last')
```

```python
##You can filter which duplicates to drop by a single column
##ddup.drop_duplicates(['col1'])
ddup.drop_duplicates(subset=['col1'])
```
